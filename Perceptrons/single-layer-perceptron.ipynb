```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
iris
```

```python
X = iris.data[: , (2,3)]
y = (iris.target == 0).astype(int)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[y==1, 0], X[y==1, 1], color="blue", marker="o", label="Setosa")
plt.scatter(X[y==0, 0], X[y==0, 1], color="green", marker="s", label="Not Setosa")

# Create a mesh grid to plot decision boundary
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                     np.linspace(y_min, y_max, 200))

plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
plt.title("Perceptron - Linear Separability of Setosa")
plt.legend()
plt.show()
```

```python
per_clf = Perceptron()
per_clf.fit(X,y)
```

```python
y_pred = per_clf.predict([[1.5,0.5]])
```
